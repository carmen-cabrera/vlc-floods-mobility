```python
def tile_to_bbox(x, y, zoom):
    """Convert tile x, y, and zoom level to bounding box in WGS84 coordinates."""
    n = 2.0 ** zoom
    lon_min = x / n * 360.0 - 180.0
    lat_min = (1.0 - (2.0 * y / n)) * 180.0 / 3.14159265358979323846
    lat_min = 180.0 / 3.14159265358979323846 * (2 * math.atan(math.exp(lat_min * 3.14159265358979323846 / 180.0)) - 3.14159265358979323846 / 2)
    
    lon_max = (x + 1) / n * 360.0 - 180.0
    lat_max = (1.0 - (2.0 * (y + 1) / n)) * 180.0 / 3.14159265358979323846
    lat_max = 180.0 / 3.14159265358979323846 * (2 * math.atan(math.exp(lat_max * 3.14159265358979323846 / 180.0)) - 3.14159265358979323846 / 2)

    return [(lon_min, lat_min), (lon_max, lat_min), (lon_max, lat_max), (lon_min, lat_max), (lon_min, lat_min)]

```
